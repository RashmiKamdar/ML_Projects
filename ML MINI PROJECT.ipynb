```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Load recipe data from CSV
recipes = pd.read_csv('D:/Recipe_Generation.csv')

# User input
user_ingredients = input("Enter ingredients separated by commas: ").split(',')

# Filter recipes based on user input
recipes['Ingredients'] = recipes['Ingredients'].astype(str)
filtered_recipes = recipes[
    recipes['Ingredients'].apply(lambda x: all(ingredient.strip() in x for ingredient in user_ingredients))
]

# Identify vegetarian and non-vegetarian recipes
def is_vegetarian(ingredients):
    non_veg_ingredients = ['chicken', 'beef', 'pork', 'fish']
    for ingredient in non_veg_ingredients:
        if ingredient in ingredients:
            return False
    return True

vegetarian_recipes = filtered_recipes[filtered_recipes['Ingredients'].apply(is_vegetarian)]
non_vegetarian_recipes = filtered_recipes[~filtered_recipes['Ingredients'].apply(is_vegetarian)]

# Display the top 3 vegetarian and non-vegetarian recipes
print("Vegetarian Recipes:")
for index, row in vegetarian_recipes[:3].iterrows():
    print(f"{row['RecipeName']} (Vegetarian)")
    print("Ingredients:")
    print(row['Ingredients'])
    print("Cooking time:")
    print(row['CookTimeInMins'])
    print("Preparation Time:")
    print(row['PrepTimeInMins'])
    print("Measurements of food according to members:")
    print(row['Servings'])
    print("Method of preparing recipe:")
    print(row['Instructions'])

print("Non-Vegetarian Recipes:")
for index, row in non_vegetarian_recipes[:3].iterrows():
    print(f"{row['RecipeName']} (Non-Vegetarian)")
    print("Ingredients:")
    print(row['Ingredients'])
    print("Cooking time:")
    print(row['CookTimeInMins'])
    print("Preparation time:")
    print(row['PrepTimeInMins'])
    print("Measurements of food according to members:")
    print(row['Servings'])
    print("Method of preparing recipe:")
    print(row['Instructions'])

# Generate bar graphs to visualize statistics
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot the number of vegetarian and non-vegetarian recipes
recipe_types = ['Vegetarian', 'Non-Vegetarian']
recipe_counts = [len(vegetarian_recipes), len(non_vegetarian_recipes)]
ax1.bar(recipe_types, recipe_counts, color=['green', 'red'])
ax1.set_title('Number of Recipes')
ax1.set_ylabel('Count')

# Plot the average cooking time for vegetarian and non-vegetarian recipes
avg_cooking_time = [
    vegetarian_recipes['CookTimeInMins'].mean(),
    non_vegetarian_recipes['CookTimeInMins'].mean()
]
ax2.bar(recipe_types, avg_cooking_time, color=['green', 'red'])
ax2.set_title('Average Cooking Time (minutes)')
ax2.set_ylabel('Time (mins)')

plt.show()

# Machine Learning Recipe Recommendation
# Use TF-IDF vectorization to convert recipe instructions to numerical features
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
recipes['Instructions'] = recipes['Instructions'].fillna('')
tfidf_matrix = tfidf_vectorizer.fit_transform(recipes['Instructions'])

# Calculate cosine similarity between recipes based on instructions
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Create a function to recommend recipes
def recommend_recipe(recipe_name, cosine_sim=cosine_sim):
    idx = recipes.index[recipes['RecipeName'] == recipe_name].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]  # Get the top 5 most similar recipes
    recipe_indices = [i[0] for i in sim_scores]
    return recipes['RecipeName'].iloc[recipe_indices]

# Ask the user for a recipe name to get recommendations
while True:
    recipe_name = input("Enter a recipe name to get recommendations (or type 'exit' to quit): ")
    if recipe_name.lower() == 'exit':
        break
    if recipe_name in recipes['RecipeName'].values:
        recommendations = recommend_recipe(recipe_name)
        print("Recommended Recipes:")
        for i, recommendation in enumerate(recommendations):
            print(f"{i + 1}. {recommendation}")
    else:
        print("Recipe not found in the dataset. Please enter a valid recipe name.")

# Ask the user if they want more recipes
want_more = input("Do you want to see more recipes? (y/n): ")

# If the user wants more recipes, display the remaining recipes
while want_more == 'y':
    print("Vegetarian Recipes:")
    for index, row in vegetarian_recipes[3:6].iterrows():
        print(f"{row['RecipeName']} (Vegetarian)")
    # Similarly, you can do the same for non-vegetarian recipes if needed
# ... (Previous code)

# User feedback and recommendation system
user_ratings = {}
while True:
    recipe_name = input("Enter a recipe name to rate (or type 'exit' to quit): ")
    
    if recipe_name.lower() == 'exit':
        break
    
    if recipe_name in recipes['RecipeName'].values:
        rating = float(input("Rate this recipe (1-5): "))
        if recipe_name in user_ratings:
            user_ratings[recipe_name].append(rating)
        else:
            user_ratings[recipe_name] = [rating]
    else:
        print("Recipe not found in the dataset. Please enter a valid recipe name.")

# Calculate recommended recipes based on user preferences
def recommend_recipe_based_on_ratings(user_ratings, cosine_sim=cosine_sim):
    # Calculate weighted cosine similarity based on user ratings
    weighted_cosine_sim = cosine_sim.copy()
    for recipe_name, ratings in user_ratings.items():
        idx = recipes.index[recipes['RecipeName'] == recipe_name].tolist()[0]
        weighted_cosine_sim[idx] = cosine_sim[idx] * (sum(ratings) / len(ratings))
    
    recommended_recipes = {}
    for recipe_name, ratings in user_ratings.items():
        idx = recipes.index[recipes['RecipeName'] == recipe_name].tolist()[0]
        sim_scores = list(enumerate(weighted_cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:6]  # Get the top 5 most similar recipes
        recommended_recipes[recipe_name] = [recipes['RecipeName'].iloc[i[0]] for i in sim_scores]

    return recommended_recipes

# Provide recipe recommendations based on user preferences
for recipe, recommendations in recommend_recipe_based_on_ratings(user_ratings).items():
    print(f"Recommended for '{recipe}':")
    for i, recommendation in enumerate(recommendations):
        print(f"{i + 1}. {recommendation}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Load recipe data from CSV
recipes = pd.read_csv('D:/Recipe_Generation.csv')

# User input
user_ingredients = input("Enter ingredients separated by commas: ").split(',')

# Filter recipes based on user input
recipes['Ingredients'] = recipes['Ingredients'].astype(str)
filtered_recipes = recipes[
    recipes['Ingredients'].apply(lambda x: all(ingredient.strip() in x for ingredient in user_ingredients))
]

# Identify vegetarian and non-vegetarian recipes
def is_vegetarian(ingredients):
    non_veg_ingredients = ['chicken', 'beef', 'pork', 'fish']
    for ingredient in non_veg_ingredients:
        if ingredient in ingredients:
            return False
    return True

vegetarian_recipes = filtered_recipes[filtered_recipes['Ingredients'].apply(is_vegetarian)]
non_vegetarian_recipes = filtered_recipes[~filtered_recipes['Ingredients'].apply(is_vegetarian)]

# Display the top 3 vegetarian and non-vegetarian recipes
print("Vegetarian Recipes:")
for index, row in vegetarian_recipes[:3].iterrows():
    print(f"{row['RecipeName']} (Vegetarian)")
    print("Ingredients:")
    print(row['Ingredients'])
    print("Cooking time:")
    print(row['CookTimeInMins'])
    print("Preparation Time:")
    print(row['PrepTimeInMins'])
    print("Measurements of food according to members:")
    print(row['Servings'])
    print("Method of preparing recipe:")
    print(row['Instructions'])

print("Non-Vegetarian Recipes:")
for index, row in non_vegetarian_recipes[:3].iterrows():
    print(f"{row['RecipeName']} (Non-Vegetarian)")
    print("Ingredients:")
    print(row['Ingredients'])
    print("Cooking time:")
    print(row['CookTimeInMins'])
    print("Preparation time:")
    print(row['PrepTimeInMins'])
    print("Measurements of food according to members:")
    print(row['Servings'])
    print("Method of preparing recipe:")
    print(row['Instructions'])

# Generate bar graphs to visualize statistics
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot the number of vegetarian and non-vegetarian recipes
recipe_types = ['Vegetarian', 'Non-Vegetarian']
recipe_counts = [len(vegetarian_recipes), len(non_vegetarian_recipes)]
ax1.bar(recipe_types, recipe_counts, color=['green', 'red'])
ax1.set_title('Number of Recipes')
ax1.set_ylabel('Count')

# Plot the average cooking time for vegetarian and non-vegetarian recipes
avg_cooking_time = [
    vegetarian_recipes['CookTimeInMins'].mean(),
    non_vegetarian_recipes['CookTimeInMins'].mean()
]
ax2.bar(recipe_types, avg_cooking_time, color=['green', 'red'])
ax2.set_title('Average Cooking Time (minutes)')
ax2.set_ylabel('Time (mins)')

plt.show()

# Machine Learning Recipe Recommendation
# Use TF-IDF vectorization to convert recipe instructions to numerical features
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
recipes['Instructions'] = recipes['Instructions'].fillna('')
tfidf_matrix = tfidf_vectorizer.fit_transform(recipes['Instructions'])

# Calculate cosine similarity between recipes based on instructions
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Create a function to recommend recipes
def recommend_recipe(recipe_name, cosine_sim=cosine_sim):
    idx = recipes.index[recipes['RecipeName'] == recipe_name].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx])
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]  # Get the top 5 most similar recipes
    recipe_indices = [i[0] for i in sim_scores]
    return recipes['RecipeName'].iloc[recipe_indices]

# Ask the user for a recipe name to get recommendations
while True:
    recipe_name = input("Enter a recipe name to get recommendations (or type 'exit' to quit): ")
    if recipe_name.lower() == 'exit':
        break
    if recipe_name in recipes['RecipeName'].values:
        recommendations = recommend_recipe(recipe_name)
        print("Recommended Recipes:")
        for i, recommendation in enumerate(recommendations):
            print(f"{i + 1}. {recommendation}")
    else:
        print("Recipe not found in the dataset. Please enter a valid recipe name.")

# Ask the user if they want more recipes
want_more = input("Do you want to see more recipes? (y/n): ")

# If the user wants more recipes, display the remaining recipes
while want_more == 'y':
    print("Vegetarian Recipes:")
    for index, row in vegetarian_recipes[3:6].iterrows():
        print(f"{row['RecipeName']} (Vegetarian)")
    # Similarly, you can do the same for non-vegetarian recipes if needed
```
